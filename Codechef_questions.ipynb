```python
from autoscraper import AutoScraper
```

```python
scraper=AutoScraper()
```

```python
url='https://www.codechef.com/problems/school?itm_medium=navmenu&itm_campaign=problems'
```

```python
wanted_list=['ATM','https://www.codechef.com/problems/HS08TEST']
```

```python
result=scraper.build(url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_9pd5':'Question','rule_60kv':'Link'})
scraper.keep_rules(['rule_9pd5','rule_60kv'])
scraper.save('Codechef')
```

```python
scraper.load('Codechef')
```

```python
result=scraper.get_result_similar(url,group_by_alias=True)
```

```python
import pandas as pd 
```

```python
df=pd.DataFrame()
```

```python
df['Question']=result['Question']
df['Link']=result['Link']
```
